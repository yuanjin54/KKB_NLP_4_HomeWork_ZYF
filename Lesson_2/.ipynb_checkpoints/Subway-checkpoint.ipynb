```python
pattern = re.compile("[\[>]([\s\S]*?)[\]<]")
def names_handler(input_list):
    temp_key = ''
    temp_list = []
    subway_connection = defaultdict(list)
    for i in input_list:
        #判断是不是line_name 是的话生成新的节点
        #if (len(i) == 3):
            s = str(i)
            #取出汉字，生成节点
            if 'subway' in s:
                res =str(pattern.findall(s))[2:-2]
                result = re.sub(',|\'|\b|\x0a','',res,10,flags=re.I)
               # print(result[3:-3].split(' ')[0])
                temp_key = result[3:-3].split(' ')[0]
                temp_list = []
                subway_connection[temp_key] = temp_list
            else:
                #print(len(i))
                s = str(i)
                if 'station' in s:
                    res =str(pattern.findall(s))[2:-2]
                    result = re.sub(',|\'|\b| ','',res,10,flags=re.I)
                    subway_connection[temp_key].append(result)
                    #print('  '+result)
                    #不是的话就取里面的汉字添加节点
    return subway_connection
result = names_handler(names_list)
```

```python
#查找一个station所属于哪一条地铁线路的方法
def find_common_stations(compare_station,isOneContain=True):
    results = {}
    line_temp_list = []
    for line in result:
        if ( compare_station in result[line]):
            line_temp_list.append(str(line))
            continue
    results[compare_station] = line_temp_list
    if isOneContain:
        return results
    else:
        if len(line_temp_list) >=2:
            return results
print(find_common_stations('垡头'))
#dic = find_common_stations('九龙山')
# for i in dic:
#     print (i)
#>>> console {'九龙山': ['7号线', '14号线']}
```

```python
def creat_change_sub_lists(input_lines_infos):
    result_line_Dict = {}
    for line_name in input_lines_infos:
        #print (line_name)
        #print(input_lines_infos[line_name])
        for stations in input_lines_infos[line_name]:
            #print (stations)
            temp_station_list = []
           # print(stations)
            #内层对比的stations
            if find_common_stations(stations,isOneContain=False) != None:
               # print(stations)
             for i in find_common_stations(stations):
                for line in find_common_stations(stations)[i]:
                     temp_station_list.append(line)
                result_line_Dict[stations] = temp_station_list
                temp_station_list = []
    return result_line_Dict
change_stations = creat_change_sub_lists(result)
print(change_stations)
```

```python
def shake_seek_lists(start_index,input_list):
    """
    只能想到这一种循环的中心四散的查找方法实现
    应该别的效率能更好，比如每次输出之后pop什么的
    但是只能想到这种了，太笨了
    start_index: the index beign seek func,
    input_list  : the list that wanna seek func
    """
    left_index = start_index-1
    right_index = start_index+1
    go_left = True
    new_seek_list = []
    for count in 2*len(input_list)*['']:#乘2防止边界碰撞长度不够
        if (go_left):
            if left_index>=0:
                new_seek_list.append(input_list[left_index])
                 #print(input_list[left_index])
                left_index -=1
                go_left = not go_left
            else:
                go_left = not go_left
        else:
            if right_index<len(input_list):
                new_seek_list.append(input_list[right_index])
                #print(input_list[right_index])
                right_index +=1
                go_left= not go_left
            else:
                go_left = not go_left
    return new_seek_list
```
